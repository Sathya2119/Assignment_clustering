```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
X=dataset.iloc[:,[3,4]].values 
```

```python
from sklearn.cluster import OPTICS
from sklearn.cluster import cluster_optics_dbscan
opt=OPTICS(min_samples=50, xi=0.05, min_cluster_size=0.05)
opt.fit(X)
```

```python
labels_150 = cluster_optics_dbscan(
    reachability=opt.reachability_,
    core_distances=opt.core_distances_,
    ordering=opt.ordering_,
    eps=0.5,
)
labels_200 = cluster_optics_dbscan(
    reachability=opt.reachability_,
    core_distances=opt.core_distances_,
    ordering=opt.ordering_,
    eps=2,
)
```

```python
space = np.arange(len(X))
reachability = opt.reachability_[opt.ordering_]
labels = opt.labels_[opt.ordering_]
```

```python
labels
```

```python
reachability
```

```python
# Reachability plot
colors = ["g.", "r.", "b.", "y.", "c."]
for klass, color in zip(range(0, 5), colors):
    Xk = space[labels == klass]
    Rk = reachability[labels == klass]
    plt.plot(Xk, Rk, color, alpha=0.3)
plt.plot(space[labels == -1], reachability[labels == -1], "k.", alpha=0.3)
plt.plot(space, np.full_like(space, 2.0, dtype=float), "k-", alpha=0.5)
plt.plot(space, np.full_like(space, 0.5, dtype=float), "k-.", alpha=0.5)
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("Reachability Plot")
```

```python
# OPTICS
colors = ["g.", "r.", "b.", "y.", "c."]
for klass, color in zip(range(0, 5), colors):
    Xk = X[opt.labels_ == klass]
    plt.plot(Xk[:, 0], Xk[:, 1], color, alpha=0.3)
plt.plot(X[opt.labels_ == -1, 0], X[opt.labels_ == -1, 1], "k+", alpha=0.1)
plt.title("Automatic Clustering\nOPTICS")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
```

```python
# DBSCAN at 0.5
colors = ["g.", "r.", "b.", "c."]
for klass, color in zip(range(0, 4), colors):
    Xk = X[labels_050 == klass]
    plt.plot(Xk[:, 0], Xk[:, 1], color, alpha=0.3)
plt.plot(X[labels_050 == -1, 0], X[labels_050 == -1, 1], "k+", alpha=0.1)
plt.title("Clustering at 0.5 epsilon cut\nDBSCAN")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")

```

```python
# DBSCAN at 2.
colors = ["g.", "m.", "y.", "c."]
for klass, color in zip(range(0, 4), colors):
    Xk = X[labels_200 == klass]
    plt.plot(Xk[:, 0], Xk[:, 1], color, alpha=0.3)
plt.plot(X[labels_200 == -1, 0], X[labels_200 == -1, 1], "k+", alpha=0.1)
plt.title("Clustering at 2.0 epsilon cut\nDBSCAN")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")

plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import OPTICS
opt=OPTICS(min_samples=50, xi=0.05, min_cluster_size=0.05)
y_opt=opt.fit_predict(X)
```
